### EDA (Exploratory data analysis)

```python
df.columns
```

```python
df1 = sns.countplot(x='Gender',data=df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in df1.containers:  #shows data labels.
    df1.bar_label(bars)
```

```python
grouping_sales = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=grouping_sales)




```

##### From above graphs, we can clearly see that women are puchasing more than men. Also women spending more amount as compare to men.

```python
df.columns
```

```python
ax = sns.countplot(x='Age Group',data = df,hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
grouping_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=grouping_age)
```

From above graphs, we can clearly see that most of the buyers are female from the age group of 26-35. 

```python
df.columns
```

```python
#Total number of orders from top 10 states.
grouping_order = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
#plt.xticks(rotation=90)
sns.set(rc={'figure.figsize':(17,7)})
sns.barplot(x='State',y='Orders',data=grouping_order)
plt.xlabel('States',size=15,color='red')
plt.ylabel('Orders',size=15,color='red')
```

```python
#total amount/sales from top 10 states
grouping_sales = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(10)
sns.barplot(x='State',y='Amount',data=grouping_sales)
sns.set(rc={'figure.figsize':(17,5)})
```

From above graph, we can see that most of the orders and sales/Amount are from Uttar Pradesh, Maharastra, Karnataka  

```python
ax = sns.countplot(x='Marital_Status',data =df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
grouping_status = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,2)})
sns.barplot(x='Marital_Status',y='Amount',data=grouping_status,hue='Gender')
```

From the above graph, we can see that most of the buyers are married women and their puchasing power is also high.

```python
ax = sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(17,7)})
plt.xlabel('Occupation',size=20,color='red')
plt.ylabel('Count',size=20,color='red')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
grouping_occupation = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(x='Occupation',y='Amount',data=grouping_occupation)
plt.xlabel('Occupation',size=20,color='red')
plt.ylabel('Amount',size=20,color='red')
```

```python
df.columns
```

From above graph we can see that most of the buyer are from IT, Healthcare and Aviation sector

```python
ax = sns.countplot(x='Product_Category',data=df,edgecolor='black')
sns.set(rc={'figure.figsize':(25,10)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel('Product_Category',color='red',size=15)
plt.ylabel('Count',color='red',size=20)
```

```python
grouping_catagory = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=grouping_catagory)
sns.set(rc={'figure.figsize':(20,5)})
```

From The above graph, we can see that most of the product sold are from Food, Clothing and then electronics. But most of the amounbt is spent on Food items.

### Final Conclusion:
The graphs clearly show that women are buying more products and spending more money than men, especially married women in the age group of 26–35 years. Most of these buyers work in sectors like IT, Healthcare, and Aviation. When we look at the locations, most of the orders and highest sales come from states such as Uttar Pradesh, Maharashtra, and Karnataka. In terms of product categories, Food, Clothing, and Electronics are the top choices. However, people spend the most money on Food items compared to other categories.
